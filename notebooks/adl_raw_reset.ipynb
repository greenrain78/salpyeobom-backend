# adl_raw_records 초기화

`adl_raw_records` 테이블의 **모든 행을 삭제**한다. 적재 노트북
(`adl_raw_ingest.ipynb`)을 실행하기 전에 먼저 이 노트북을 실행해 테이블을 비운다.

⚠️ 이 작업은 되돌릴 수 없다. 스키마(테이블·컬럼)는 유지되고 행 데이터만 삭제된다.

```python
import os
import sys
from pathlib import Path

# 작업 디렉토리를 프로젝트 루트로 이동(.env.local·app import 해석용, 폴더 생성 아님).
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
os.chdir(ROOT)
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

print(f"프로젝트 루트: {ROOT}")
```

```python
from tortoise import Tortoise

from app.database import TORTOISE_ORM
from app.models.adl_raw import AdlRawRecord

if not Tortoise._inited:
    await Tortoise.init(config=TORTOISE_ORM)
await Tortoise.generate_schemas()
print("DB 연결 완료")
```

```python
before = await AdlRawRecord.all().count()
deleted = await AdlRawRecord.all().delete()
after = await AdlRawRecord.all().count()
print(f"삭제 전 {before}건 → 삭제 {deleted}건 → 삭제 후 {after}건")
```

```python
await Tortoise.close_connections()
print("DB 연결 종료")
```
